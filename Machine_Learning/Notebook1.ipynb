```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('D:\Lexicon_Labs\Machine_Learning\Files\house_price.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df_filtered = df[df['bedrooms'] > 2]
print(df_filtered)
```

```python
sorted_df = df.sort_values(by='sales_price', ascending=True)
print(sorted_df)
```

```python
sale_average = df['sales_price'].mean()
print(sale_average)
```

```python
df['price_per_sqmtr'] = df['sales_price'] / df['sq_meter']
df.head()
```

```python
avg_by_neighborhood = df.groupby('neighborhood')['sales_price'].mean()
print(avg_by_neighborhood)
```

```python
avg_by_neighborhood.plot(kind='bar', color='skyblue')

import matplotlib.pyplot as plt
plt.xlabel('Neighborhood')
plt.ylabel('Average Sales Price')
plt.title('Average Sales Price per Neighborhood')
plt.xticks(rotation=45)
plt.show()
```
