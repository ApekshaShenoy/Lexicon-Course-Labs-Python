```python
#Convert age to numeric
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
#unrealistic ages
df.loc[(df['age'] < 0) | (df['age'] > 120), 'age'] = pd.NA
```

```python
#Fill missing age with median
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)
```

```python
#Height must be positive

#Fix missing and placeholder values

#Decisions

#-99 is an invalid placeholder

#Height ≤ 0 → missing

#Missing → fill with median height

df['height'] = pd.to_numeric(df['height'], errors='coerce') #Convert to numeric
df.loc[df['height'] <= 0, 'height'] = pd.NA   #Remove invalid heights
#Fill missing height
median_height = df['height'].median()
df['height'] = df['height'].fillna(median_height)
df['height'].describe()
```

```python
# Handle missing income

# Decide what to do with 0

# Decision (simple and defensible)

# Blank income → missing

# 0 → treated as missing (likely placeholder)

# Fill with median income

#Convert to numeric
df['income'] = pd.to_numeric(df['income'], errors='coerce')

#Treat 0 as missing
df.loc[df['income'] == 0, 'income'] = pd.NA

#Fill missing income
median_income = df['income'].median()
df['income'] = df['income'].fillna(median_income)

df['income'].describe()
```

```python
def clean_purchases(x):
    if pd.isna(x) or x == "":
        return [0]
    items = x.split(',')
    cleaned = []
    for item in items:
        if item.strip() == "":
            cleaned.append(0)
        else:
            cleaned.append(int(item))
    return cleaned
```

```python
df['purchases'] = df['purchases'].apply(clean_purchases)
df['purchases'].head()
```

```python
df = pd.get_dummies(df, columns=['gender'])
```

```python
cols_to_normalize = ['height', 'weight', 'income']

for col in cols_to_normalize:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
```

```python
df.to_csv("D:\Lexicon_Labs\Machine_Learning\Files_Day_2\cleaned_data.csv", index=False)
```
